```python
import pandas as pd
import seaborn as sp
import matplotlib.pyplot as plt

car_dataset=pd.read_csv('train-data.csv')

```

```python
import warnings 

warnings.filterwarnings('ignore')
```

```python
car_dataset
```

```python
car_dataset.info()
```

```python
car_dataset.isnull().sum()
```

```python
car_dataset.drop(["Unnamed: 0","New_Price"],axis=1,inplace=True)
```

```python
car_dataset
```

```python
car_dataset=car_dataset.dropna()
car_dataset
```

```python
car_dataset=car_dataset.reset_index(drop=True)
car_dataset
```

```python
for i in range(car_dataset.shape[0]):
    car_dataset.at[i,'Company']=car_dataset['Name'][i].split()[0]
    car_dataset.at[i,'Mileage_kg/km']=car_dataset['Mileage'][i].split()[0]
    car_dataset.at[i,'Engine_cc']=car_dataset['Engine'][i].split()[0]
    car_dataset.at[i,'Power_bhp']=car_dataset['Power'][i].split()[0]


```

```python
car_dataset
```

```python
car_dataset.drop("Name",axis=1,inplace=True)
```

```python
car_dataset.drop("Mileage",axis=1,inplace=True)
car_dataset.drop("Engine",axis=1,inplace=True)
car_dataset.drop("Power",axis=1,inplace=True)
```

```python
car_dataset
```

```python
car_dataset['Mileage_kg/km']=car_dataset['Mileage_kg/km'].astype(float)
car_dataset['Engine_cc']=car_dataset['Engine_cc'].astype(float)
```

```python
car_dataset[car_dataset['Power_bhp']=='null']
```

```python
count=0
position=[]
for i in range(car_dataset.shape[0]):
    if car_dataset['Power_bhp'][i]=='null':
        count =count+1
        position.append(i)
        
        
```

```python
car_dataset=car_dataset.drop(car_dataset.index[position])

car_dataset=car_dataset.reset_index(drop=True)
```

```python
car_dataset
```

```python
car_dataset['Power_bhp']=car_dataset['Power_bhp'].astype(float)
```

```python
car_dataset.Location.value_counts()
```

```python
sp.catplot(y="Price",x="Location",data=car_dataset.sort_values('Price',ascending=False),kind="bar",height=6,aspect=3)
```

```python
Location=car_dataset[['Location']]
Location=pd.get_dummies(Location)
Location
```

```python
car_dataset.Fuel_Type.value_counts()
```

```python
Fuel_Types=car_dataset[['Fuel_Type']]
Fuel_Types=pd.get_dummies(Fuel_Types)
Fuel_Types
```

```python
Transmission=car_dataset[['Transmission']]
Transmission=pd.get_dummies(Transmission,drop_first=True)
Transmission
```

```python
car_dataset.Owner_Type.value_counts()
```

```python
car_dataset['Owner_Type'].replace({"First":4,"Second":3,"Third":2,"Fourth & Above":1},inplace=True)
car_dataset
```

```python
car_dataset.drop('Company',axis=1,inplace=True)
```

```python
final_train=pd.concat([car_dataset,Location,Fuel_Types,Transmission],axis=1,inplace=True)
```

```python
final_train
```

```python
final_train.drop(['Location','Fuel_Type','Transmission'],axis=1,inplace=True)
```

```python
final_train
```

```python
price=final_train['Price']
final_train.drop(['Price'],axis=1,inplace=True)
```

```python
final_train['Price']=price
```

```python
final_train
```

```python
sp.pairplot(car_dataset)
```

```python
plt.figure(figsize=(18,18))

sp.heatmap(final_train.corr(),annot=True,cmap="RdYlGn")
plt.show()
```

```python
X=final_train.drop('Price',axis=1)
X
```

```python
y=final_train['Price']
y
```

```python
from sklearn.tree import ExtraTreeRegressor

selection=ExtraTreeRegressor()
selection.fit(X,y)
```

```python
plt.figure(figsize=(12,18))

feat_importances=pd.Series(selection.feature_importances_,index=X.columns)
feat_importances.nlargest(20).plot(kind="barh")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

linerar_reg=LinearRegression()
linear=linerar_reg.fit(X_train,y_train)
y_predic=linerar_reg.predict(x_test)
print(linerar_reg.score(X_train,y_train))
```

```python
print(linerar_reg.score(x_test,y_test))
```

```python
plt.scatter(y_test,y_predic)
```

```python
from sklearn import linear_model
lasso_reg=linear_model.Lasso(alpha=50,tol=0.1)
lasso_reg.fit(X_train,y_train)
```

```python
lasso_reg.score(x_test,y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor

df_reg=DecisionTreeRegressor()
decision=df_reg.fit(X_train,y_train)
y_predic=df_reg.predict(x_test)
print(df_reg.score(X_train,y_train))
print(df_reg.score(x_test,y_test))
```

```python
from sklearn.model_selection import cross_val_score
result=cross_val_score(DecisionTreeRegressor(splitter='best',min_samples_leaf=2,max_depth=15),X,y,cv=5)
result
```

```python
from sklearn.ensemble import RandomForestRegressor

rdf_reg=RandomForestRegressor()
randoms=rdf_reg.fit(X_train,y_train)
y_predic=rdf_reg.predict(x_test)
print(df_reg.score(X_train,y_train))
print(df_reg.score(x_test,y_test))
```
